```python
import pandas as pd
```

## Load Datasets

```python
lyrics_data = pd.read_csv("/content/drive/MyDrive/SEM4/Natural Language Processing/Project NLP/Project - NLP shared/Data Preprocessing/lyrics-data.csv")

print(lyrics_data.shape)
lyrics_data.head()
```

```python
artist_data = pd.read_csv("/content/drive/MyDrive/SEM4/Natural Language Processing/Project NLP/Project - NLP shared/Data Preprocessing/artists-data.csv")

print(artist_data.shape)
artist_data.head()
```

## Merge lyrics data and artists data

```python
merged_df = lyrics_data.merge(artist_data, left_on='ALink', right_on='Link', how='left')

print(merged_df.shape)
merged_df.head()
```

## Remove any non-english lyrics, and drop all columns but lyric and genres

```python
data_en = merged_df[merged_df['language'] == 'en']

data_en = data_en.drop(['ALink', 'SName', 'SLink', 'Artist', 'language', 'Songs', 'Popularity', 'Link'], axis=1)

print(data_en.shape)
data_en.head()
```

```python
data_en_filtered = pd.read_csv("/content/drive/MyDrive/SEM4/Natural Language Processing/Project NLP/Project - NLP shared/Data Preprocessing/filtered_lyrics_two.csv")

print(data_en_filtered.shape)
data_en_filtered.head()
```

## Make every row only contains a single genre

```python
data_en_single_genre = data_en.assign(Genres=data_en['Genres'].str.split('; ')).explode('Genres').reset_index(drop=True)

print(data_en_single_genre.shape)
data_en_single_genre.head()
```

```python
LYRICS_GENRE = 'Folk'
```

```python
en_genre_lyrics = data_en_single_genre[data_en_single_genre['Genres'] == LYRICS_GENRE]

print(en_genre_lyrics.shape)
en_genre_lyrics.head(10)
```
